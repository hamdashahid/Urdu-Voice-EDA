## <font face="Gramond" color="orange">**Task 03**</font>


#### <font face="Gramond" color="white">_Sentence Length Analysis_</font>

<hr style="border: 1px solid orange;"> 


* #### <font face="Gramond" color="white">_Length_</font>

```python
# q3 code \
sentences = dev['sentence']
# dev['sentence'][0].count('')
# sentences
dic = {'sentence' : [],'length' : []}
for i in range (len(sentences)):
#    print(sentences[i],sentences[i].count(' '))
    dic['sentence'].append(sentences[i])
    dic['length'].append(len(sentences[i]))
df = pd.DataFrame(dic)
dev['sentence_length'] = df['length']
df    
```

* #### <font face="Gramond" color="white">_Plot_</font>

```python
# df.plot(df['length'].value_counts(),kind='bar', title='Sentence Length Distribution')
df['length'].value_counts().sort_index().plot(kind='bar', title='Sentence Length Distribution',figsize=(20,5))
plt.legend()
```

```python
avg_by_gender = dev.groupby('gender')['sentence_length'].mean()
avg_by_age = dev.groupby('age')['sentence_length'].mean()

avg_by_gender.plot(kind='bar', title='Average Sentence length by gender')
plt.legend()
```

```python
avg_by_age.plot(kind='bar', title='Average Sentence length by Age') 
plt.legend()
```

* #### <font face="Gramond" color="white">_Vilon plot_</font>

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 15))
sns.violinplot(x='age', y='sentence_length', data=dev,palette='coolwarm')
plt.title('Sentence Length Distribution by Age')
plt.show()
```

## <font face="Gramond" color="orange">**Task 04**</font>




#### <font face="Gramond" color="white">_Audio duration Analysis_</font>

<hr style="border: 1px solid orange;"> 


* #### <font face="Gramond" color="white">_Duration Count_</font>


```python
# q4 code

import librosa as lib
dict = {'path': dev['path'],'durattion':[]}
for i in dev['path']:
    audio_path = f'cv-corpus-19.0-2024-09-13/ur/clips/{i}'
    y, sr = lib.load(audio_path)

    duration = lib.get_duration(y=y, sr=sr)
    dict['durattion'].append(duration)
df3 = pd.DataFrame(dict)
dev['duration'] = df3['durattion']
dev
```

* #### <font face="Gramond" color="white">_Plot of audio durations_</font>


```python
dev['duration'].value_counts().sort_index().plot(kind='bar', title='Audio Duration Distribution',figsize=(50,10))
plt.legend()
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
```

* #### <font face="Gramond" color="white">_Audio durations by gender_</font>


```python
avg_duration_by_gender = dev.groupby('gender')['duration'].mean()
avg_duration_by_age = dev.groupby('age')['duration'].mean()

avg_duration_by_gender.plot(kind='bar', title='Average Audio Duration by Gender')
plt.legend()
```

* #### <font face="Gramond" color="white">_Audio durations by age_</font>


```python
avg_duration_by_age.plot(kind='bar', title='Average Audio Duration by Age')
plt.legend()
```

* #### <font face="Gramond" color="white">_Outliers_</font>


```python
dev= dev[(dev['duration'] < 10) & (dev['duration'] > 1)]
# dev['duration']
# dev['duration'].mean()
dev['duration'].value_counts().sort_index().plot(kind='bar', title='Revised Audio Duration Distribution')
plt.legend()
```

```python
avg_duration_by_gender = dev.groupby('gender')['duration'].mean()
avg_duration_by_age = dev.groupby('age')['duration'].mean()

avg_duration_by_gender.plot(kind='bar', title='Average Audio Duration by Gender')
plt.legend()
```

```python
avg_duration_by_age.plot(kind='bar', title='Average Audio Duration by Age')
plt.legend()
```
